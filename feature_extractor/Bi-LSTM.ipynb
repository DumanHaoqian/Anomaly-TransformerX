# Bi-directional Long-Short Term Memory

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import recall_score, accuracy_score, f1_score, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
from tqdm import tqdm

import os
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

train_data = np.load('/home/haoqian/anomaly/SMD/SMD_train.npy')
test_data = np.load('/home/haoqian/anomaly/SMD/SMD_test.npy')
test_label = np.load('/home/haoqian/anomaly/SMD/SMD_test_label.npy')

def preprocess_data(data, scaler=None):
    if scaler is None:
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(data)
    else:
        data_scaled = scaler.transform(data)
    
    df = pd.DataFrame(data_scaled)
    df.fillna(method='ffill', inplace=True)
    df.fillna(method='bfill', inplace=True)
    return df.values, scaler

train_data_processed, scaler = preprocess_data(train_data)
test_data_processed, _ = preprocess_data(test_data, scaler)

window_size = 10
def create_sequences(data, window_size=10):
    sequences = []
    for i in range(len(data) - window_size):
        sequences.append(data[i:i + window_size])
    return np.array(sequences)

X_train = create_sequences(train_data_processed, window_size)
X_test = create_sequences(test_data_processed, window_size)
y_test = test_label[window_size:]

class BiLSTMAutoencoder(nn.Module):
    def __init__(self, input_size, hidden_size=32):
        super().__init__()
        self.encoder = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size // 2,
            num_layers=1,
            batch_first=True,
            bidirectional=True
        )
        self.decoder = nn.LSTM(
            input_size=hidden_size,
            hidden_size=hidden_size,
            num_layers=1,
            batch_first=True
        )
        self.output_layer = nn.Linear(hidden_size, input_size)
        
    def forward(self, x):
        _, (hidden, cell) = self.encoder(x)
        hidden_forward, hidden_backward = hidden[0], hidden[1]
        hidden_concat = torch.cat([hidden_forward, hidden_backward], dim=1).unsqueeze(0)
        cell_forward, cell_backward = cell[0], cell[1]
        cell_concat = torch.cat([cell_forward, cell_backward], dim=1).unsqueeze(0)
        decoder_input = torch.zeros(x.size(0), x.size(1), self.decoder.input_size).to(x.device)
        output, _ = self.decoder(decoder_input, (hidden_concat, cell_concat))
        return self.output_layer(output)

model = BiLSTMAutoencoder(input_size=X_train.shape[2]).to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

loss_history = []

def train_in_batches(model, train_loader, epochs=50):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        for batch_x, _ in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
            optimizer.zero_grad()
            reconstructions = model(batch_x)
            loss = criterion(reconstructions, batch_x)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            del reconstructions, loss
            torch.cuda.empty_cache()
        avg_loss = total_loss / len(train_loader)
        loss_history.append(avg_loss)  # 记录损失
        print(f"Epoch {epoch+1}, Loss: {avg_loss:.4f}")
    return model


batch_size = 64
train_dataset = TensorDataset(
    torch.FloatTensor(X_train).to(device), 
    torch.FloatTensor(X_train).to(device)
)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

model = train_in_batches(model, train_loader)

plt.figure(figsize=(10, 5))
plt.plot(loss_history, label='Training Loss')
plt.title('Training Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

def compute_scores_safely(model, data, batch_size=64):
    model.eval()
    scores = []
    with torch.no_grad():
        for i in range(0, len(data), batch_size):
            batch = torch.FloatTensor(data[i:i + batch_size]).to(device)
            reconstructions = model(batch)
            mse = torch.mean((batch - reconstructions) ** 2, dim=(1, 2))
            scores.extend(mse.cpu().numpy())
            del batch, reconstructions, mse
            torch.cuda.empty_cache()
    return np.array(scores)

train_scores = compute_scores_safely(model, X_train)
test_scores = compute_scores_safely(model, X_test)

threshold = np.quantile(train_scores, 0.99)
test_preds = (test_scores > threshold).astype(int)

def evaluate_results(true_labels, pred_labels, anomaly_scores):
    print("\nEvaluation Metrics:")
    print(f"Accuracy: {accuracy_score(true_labels, pred_labels):.4f}")
    print(f"F1 Score: {f1_score(true_labels, pred_labels):.4f}")
    print(f"Recall: {recall_score(true_labels, pred_labels):.4f}")
    print(f"ROC AUC: {roc_auc_score(true_labels, anomaly_scores):.4f}")
    print("\nConfusion Matrix:")
    print(confusion_matrix(true_labels, pred_labels))

    plt.figure(figsize=(12, 6))
    plt.plot(anomaly_scores, label='Anomaly Scores')
    plt.scatter(np.where(true_labels == 1)[0], 
                anomaly_scores[true_labels == 1], 
                color='red', label='True Anomalies')
    plt.axhline(threshold, color='g', linestyle='--', label='Threshold')
    plt.legend()
    plt.show()

evaluate_results(y_test, test_preds, test_scores)

def extract_features(model, data, batch_size=64):
    model.eval()
    features = []
    with torch.no_grad():
        for i in range(0, len(data), batch_size):
            batch = torch.FloatTensor(data[i:i + batch_size]).to(device)
            _, (hidden, _) = model.encoder(batch)
            hidden = hidden.permute(1, 0, 2)
            hidden = hidden.reshape(hidden.size(0), -1)
            features.extend(hidden.cpu().numpy())
            del batch, hidden
            torch.cuda.empty_cache()
    return np.array(features)

train_features = extract_features(model, X_train)
test_features = extract_features(model, X_test)

from sklearn.ensemble import IsolationForest
clf = IsolationForest(contamination=0.05).fit(train_features)
test_preds = clf.predict(test_features)
test_preds = (test_preds == -1).astype(int)
evaluate_results(y_test, test_preds, clf.decision_function(test_features))

from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

pca = PCA(n_components=3)
vis_features = pca.fit_transform(test_features)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(vis_features[:, 0], vis_features[:, 1], vis_features[:, 2], 
                 c=y_test, alpha=0.5, cmap='coolwarm',s=10)

plt.title("3D PCA of BiLSTM Features (Red=Anomaly)")
plt.colorbar(sc)
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
def extend_features_with_regression(train_features, test_features, lookback=50, forecast=10):
    # Initialize extended arrays
    extended_train = np.zeros((train_features.shape[0] + forecast, train_features.shape[1]))
    extended_test = np.zeros((test_features.shape[0] + forecast, test_features.shape[1]))
    
    # Copy original data
    extended_train[:train_features.shape[0]] = train_features
    extended_test[:test_features.shape[0]] = test_features
    
    # For each feature dimension
    for dim in range(train_features.shape[1]):
        # Prepare training data (last 'lookback' points from train)
        X_train = np.arange(lookback).reshape(-1, 1)
        y_train = train_features[-lookback:, dim]
        
        # Train linear regression model
        model = LinearRegression()
        model.fit(X_train, y_train)
        
        # Predict next 'forecast' points for train
        X_pred = np.arange(lookback, lookback + forecast).reshape(-1, 1)
        train_predictions = model.predict(X_pred)
        extended_train[train_features.shape[0]:, dim] = train_predictions
        
        # Prepare test data (last 'lookback' points from test)
        X_test = np.arange(lookback).reshape(-1, 1)
        y_test = test_features[-lookback:, dim]
        
        # Retrain model on test data (or you could use the same model)
        model_test = LinearRegression()
        model_test.fit(X_test, y_test)
        
        # Predict next 'forecast' points for test
        test_predictions = model_test.predict(X_pred)
        extended_test[test_features.shape[0]:, dim] = test_predictions
    
    return extended_train, extended_test

# Print original shapes
print("Original shapes:")
print(f"Train features: {train_features.shape}")  # (708395, 32)
print(f"Test features: {test_features.shape}")   # (708410, 32)

# Extend features with regression
extended_train, extended_test = extend_features_with_regression(
    train_features, test_features, lookback=50, forecast=10
)

# Print extended shapes
print("\nExtended shapes:")
print(f"Extended train features: {extended_train.shape}")  # (708405, 32)
print(f"Extended test features: {extended_test.shape}")    # (708420, 32)

train_feature=np.save("/home/haoqian/anomaly/features/Lstm/lstm_train_features.npy",extended_train)
test_feature=np.save("/home/haoqian/anomaly/features/Lstm/lstm_test_features.npy",extended_test)
```

```python
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, average_precision_score

def performance_measurement(test_label, pred_label):
    precision = precision_score(test_label, pred_label, zero_division=0)
    recall = recall_score(test_label, pred_label, zero_division=0)
    f1 = f1_score(test_label, pred_label, zero_division=0)
    try:
        auc_roc = roc_auc_score(test_label, pred_label)
    except ValueError:
        auc_roc = 0.0 
    return [
        round(precision, 4),
        round(recall, 4),
        round(f1, 4),
        round(auc_roc, 4)
    ]

train_feature=np.load("/home/haoqian/anomaly/features/Lstm/lstm_train_features.npy")
test_feature=np.load("/home/haoqian/anomaly/features/Lstm/lstm_test_features.npy")
print(train_feature.shape)
print(test_feature.shape)
train_data=np.load("/home/haoqian/anomaly/SMD/SMD_train.npy")
test_data=np.load("/home/haoqian/anomaly/SMD/SMD_test.npy")
test_label=np.load("/home/haoqian/anomaly/SMD/SMD_test_label.npy")


input_data_train=np.hstack([train_data,train_feature])
input_data_test=np.hstack([test_data,test_feature])

print("="*80)

print("Feature Fusion dimention:")

print(input_data_train.shape)
print(input_data_test.shape)


print("="*80)
print("Isolation Forest on Original Data")
print("="*80)


if_original = IsolationForest(n_estimators=200, 
                             max_samples=256, 
                             contamination='auto', 
                             random_state=42)
if_original.fit(train_data)

original_scores = -if_original.score_samples(test_data) 
original_pred = (original_scores > np.quantile(original_scores, 0.95)).astype(int) 

metrics_original = performance_measurement(test_label, original_pred)
print(f"Precision: {metrics_original[0]}")
print(f"Recall: {metrics_original[1]}")
print(f"F1-score: {metrics_original[2]}")
print(f"AUC-ROC: {metrics_original[3]}")

print("\n" + "="*80)
print("Isolation Forest on Extracted Features")
print("="*80)


if_features = IsolationForest(n_estimators=200, 
                            max_samples=256, 
                            contamination='auto', 
                            random_state=42)
if_features.fit(train_feature)

features_scores = -if_features.score_samples(test_feature)
feature_pred = (features_scores > np.quantile(features_scores, 0.95)).astype(int)

metrics_features = performance_measurement(test_label, feature_pred)
print(f"Precision: {metrics_features[0]}")
print(f"Recall: {metrics_features[1]}")
print(f"F1-score: {metrics_features[2]}")
print(f"AUC-ROC: {metrics_features[3]}")

print("\n" + "="*80)
print("Isolation Forest on Extracted Features and Original Data")
print("="*80)


if_comb = IsolationForest(n_estimators=200, 
                            max_samples=256, 
                            contamination='auto', 
                            random_state=42)
if_comb.fit(input_data_train)

comb_scores = -if_comb.score_samples(input_data_test)
comb_pred = (comb_scores > np.quantile(comb_scores, 0.95)).astype(int)

# 评估指标
metrics_comb = performance_measurement(test_label, comb_pred)
print(f"Precision: {metrics_comb[0]}")
print(f"Recall: {metrics_comb[1]}")
print(f"F1-score: {metrics_comb[2]}")
print(f"AUC-ROC: {metrics_comb[3]}")

```
